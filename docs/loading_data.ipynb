```python
import pandas as pd
import numpy
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split

iowa_file_path = './DataSets/home-data-for-ml-course_train.csv'
```

```python
home_data = pd.read_csv(iowa_file_path)
```

```python
home_data.describe()
```

```python
home_data.columns

```

```python
# home_data = home_data.dropna(axis=0)
# home_data.columns
```

```python
# home_data.Price
y = home_data.SalePrice
feature_names = ['LotArea',
                 'YearBuilt',
                 '1stFlrSF',
                 '2ndFlrSF',
                 'FullBath',
                 'BedroomAbvGr',
                 'TotRmsAbvGrd'
                 ]
```

```python
X = home_data[feature_names]
```

```python
X.describe()
```

```python
X.head()
```

```python
iowa_model = DecisionTreeRegressor(random_state=1)
```

```python
iowa_model.fit(X, y)
```

```python
salePrices = iowa_model.predict(X)
```

```python
dictionary = dict(zip(y, salePrices.tolist()))
```

```python
print( {k:v for (k,v) in dictionary.values().items() if abs(k - v) > 0}) 
```

```python
mean_absolute_error(y, salePrices)
```

```python
# split data into training and validation data, for both features and target
# The split is based on a random number generator. Supplying a numeric value to
# the random_state argument guarantees we get the same split every time we
# run this script.
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state = 1)
```

```python
# Define model
melbourne_model = DecisionTreeRegressor()
# Fit model
melbourne_model.fit(train_X, train_y)
```

```python
# get predicted prices on validation data
val_predictions = melbourne_model.predict(val_X)
print(mean_absolute_error(val_y, val_predictions))
```
